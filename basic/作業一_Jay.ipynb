```python
import tensorflow as tf
import numpy as np
from tfdot import tfdot
tf.reset_default_graph()

s = tf.constant(0)
for i in range(11):
    s=s+i
sess=tf.Session()
sess.run(s)
tfdot()
```

```python
tf.reset_default_graph()
sess=tf.Session()
i = tf.constant(np.arange(11),name='test')
i_sum= tf.reduce_sum(i,name='sum')
sess.run(i_sum)
```
